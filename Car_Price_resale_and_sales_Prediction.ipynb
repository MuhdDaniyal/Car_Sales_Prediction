Missing Data

```python
#  Handling the null values

print("\n1- Do you want to drop null values?\n2- Do you want to fill null values with median?")
op1 = int(input("Enter: "))
if op1 == 1:
    dataset = dataset.dropna()

elif op1 == 2:
    dataset.__year_resale_value = dataset.__year_resale_value.fillna(dataset.__year_resale_value.median())
    dataset.Price_in_thousands = dataset.Price_in_thousands.fillna(dataset.Price_in_thousands.median())
    dataset.Engine_size = dataset.Engine_size.fillna(dataset.Engine_size.median())
    dataset.Horsepower = dataset.Horsepower.fillna(dataset.Horsepower.median())
    dataset.Wheelbase = dataset.Wheelbase.fillna(dataset.Wheelbase.median())
    dataset.Width = dataset.Width.fillna(dataset.Width.median())
    dataset.Length = dataset.Length.fillna(dataset.Length.median())
    dataset.Curb_weight = dataset.Curb_weight.fillna(dataset.Curb_weight.median())
    dataset.Fuel_capacity = dataset.Fuel_capacity.fillna(dataset.Fuel_capacity.median())
    dataset.Fuel_efficiency = dataset.Fuel_efficiency.fillna(dataset.Fuel_efficiency.median())
    dataset.Power_perf_factor = dataset.Power_perf_factor.fillna(dataset.Power_perf_factor.median())

else :
    print("Error!!! Invalid input. Please try again.")

dataset.info()

# dataset.head(50)
```

Detecting Outliers

```python
# Detecting outliers through function IQR method. Later we'll see the outliers through visualization and may be remove the outliers if accuracy is not good
def identify_outliers(dataset):
    Q1 = np.percentile(dataset, 25)
    Q3 = np.percentile(dataset, 75)
    IQR = Q3 - Q1
    cutoff = 1.5 * IQR
    lower, upper = Q1 - cutoff, Q3 + cutoff
    outliers = [i for i in dataset if i < lower or i > upper]
    return outliers

print("\n----------------------------------------------------------------------------------------------------------------\n")
print("Outliers in Sales_in_thousands\n",identify_outliers(dataset.iloc[:,2]))
print("\n----------------------------------------------------------------------------------------------------------------\n")
print("Outliers in __year_resale_value\n",identify_outliers(dataset.iloc[:,3]))
print("\n----------------------------------------------------------------------------------------------------------------\n")
print("Outliers in Price_in_thousandss\n",identify_outliers(dataset.iloc[:,5]))
print("\n----------------------------------------------------------------------------------------------------------------\n")

# 2 -> Sales_in_thousands
# 3 -> __year_resale_value
# 5 -> Price_in_thousands
```
